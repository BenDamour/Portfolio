# Data cleaning

Suppose that we want to have a representative test_set of the global population. We have to create category for important feature, to be representative. Here : median income is an important feature, so we need to make categories.

```python
housing["median_income"].hist()
```

```python
housing["income_cat"]=pd.cut(housing["median_income"],bins=[0.,1.5,3.0,4.5,6.,np.inf], labels=[1,2,3,4,5])
housing['income_cat'].hist()
```

We can now try to make a representative test and train set for ou ML

```python
# We have to split the data set before making some data snooping bias
import numpy as np 

def split_train_test(data,test_ratio):
    """
    Given a Data Frame, split the complete data set in two parts : a train_set and 
    a test_set. 
    The correponsding test_set contains a ratio (in %) of the total data in the data set.
    
    Return a tuple train_set,test_set
   
    """
    np.random.seed(42)
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]

train_set, test_set = split_train_test(housing,0.2)


# or 

from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(housing,test_size=0.2,random_state=42)

    
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index,test_index in split.split(housing,housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]

# We can compare the reprensentativity of our stratified shuffle VS our random
print('Overall : Housing')
print(housing['income_cat'].value_counts().sort_index()/len(housing))

print('Our random test set')
print(test_set['income_cat'].value_counts().sort_index()/len(test_set))

print('Our stratified random set')
print(strat_test_set['income_cat'].value_counts().sort_index()/len(test_set))
```

We have now our strat_train_set and our strat_test_set, which is representative of our population (based on the median_income)

We can remove the income_cat attribute from our dataframe:

    

```python
for set_ in (strat_train_set,strat_test_set):
    set_.drop('income_cat',axis="columns",inplace=True)
```

## Custom transformers

Allow you tu custom your cleanup operations or combining attributes.

We have to build our own class and implement three methods: fit(), transform(), fit_tranform()

Here an example for adding the combined attributes we discuss above :


```python
from sklearn.base import BaseEstimator, TransformerMixin 
# without (*args and **kargs) BaseEstimator give methods get_params() and set_params()
# TransfromerMixin build automaticaly fit_transform() (if fit and transform is defined)

rooms_ix, bedrooms_ix, population_ix, households_ix = 3,4,5,6

class CombineAttributesAdder(BaseEstimator,TransformerMixin):
    
    def __init__(self,add_berooms_per_room = True):
        self.add_berooms_per_room = add_berooms_per_room
        
    def fit(self,X,y=None):
        return self

    def transform(self,X):
        rooms_per_household = X[:,rooms_ix] / X[:,households_ix]
        population_per_household=X[:,population_ix]/X[:,households_ix]
        if self.add_berooms_per_room:
            bedrooms_per_room = X[:,bedrooms_ix]/X[:,rooms_ix]
            #np.c_ : concatenation along the second axis
            return np.c_[X,rooms_per_household,population_per_household,bedrooms_per_room]
        else:
            return np.c_[X,rooms_per_household,population_per_household]


# We can choose to add our feature attirbute or not.

hyper_parameter_add_bedrooms = True 

attr_adder = CombineAttributesAdder(add_berooms_per_room=hyper_parameter_add_bedrooms)

housing_extra_attribs = attr_adder.fit_transform(housing.values)

# Transform our np_array in Data_Frames
if hyper_parameter_add_bedrooms:
    test = pd.DataFrame(housing_extra_attribs,columns=(list(housing.columns)+["rooms_per_household","population_per_household","bedrooms_per_room"]),index=housing.index) 
else: 
    test = pd.DataFrame(housing_extra_attribs,columns=(list(housing.columns)+["rooms_per_household","population_per_household"]),index=housing.index) 
test.head()
```

## Feature Scaling 

ML algorithm's don't perform well with different scales. 

Solution to get the same scale:
1. Normalization (min-max scaling): Values are shifted and rescaled form 0 to 1. (value - min)/(max-min)    OR We can use the MinMaxScaler of sklearn with feature_range hyperparameter, if we want to change the range.

2. Standardization : Substract the mean value (standardize values always have a zero mean) and we divides by the standard deviation so the resulting distribution has unit variance. OR use the StandardScaler of sklearn!! WARNING !! The values are not bounded, not good for neural networks 


### NOTE : Standardisation is much less affected by outliers ! 

### NOTE2 : It's important to fit the scalers to the training data only !! Only then can we use them to transform the training set and the test set (and new data)

## Transformation pipeline

Pipeline calss help us to make the data transformation following the right steps of transformation !

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler #To make standardization

num_pipeline = Pipeline([('imputer', SimpleImputer(strategy='median')),
                         ('attribs_adder', CombineAttributesAdder()),
                         ('std_scaler',StandardScaler())
                        ])


housing_num_tr = num_pipeline.fit_transform(housing_num)
```

#### The best transformation pipeline 

A single transformer to handle all the columns of our data and apply the appropriate transformations : With ColumnTransformer from sklearn.

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
    ("num",num_pipeline, num_attribs),
    ("cat",OneHotEncoder(), cat_attribs),    
])

housing_prepared = full_pipeline.fit_transform(housing)
```

## Training and Evaluating on the Training Set

### Linear Regression model

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_prepared,housing_lables)
```

The model is trained, we have to check the prediction and the lables. 

```python
some_data = housing.iloc[:5]
some_labels = housing_lables.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)
print("Predictions:", lin_reg.predict(some_data_prepared))
print("Lables:", list(some_labels))

# Here print the difference in %, between reality and prediction
print((list(some_labels)/lin_reg.predict(some_data_prepared)-1)*100)
```

We can mesure the RMSE (typical prediction error)

```python
from sklearn.metrics import mean_squared_error

housing_prediction = lin_reg.predict(housing_prepared)

# Calculate the RMSE on the whole training set
lin_mse = mean_squared_error(housing_lables,housing_prediction)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

The error is too huge. With this error with a median house price between \\$120,000 and \\$265,000 a prediction error arround \\$68,628 is huge.

We underfit our model !
Why : 
1. Some features does not provide enough information to make good predictions
2. The model is too powerless 

Solution: 
1. Reduce the constraints on the model (but here we have no hyperparameter, it's not regularized)
2. Try a complex model 



### Decision Tree Regressor (non linear relationship) model 

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(housing_prepared,housing_lables)
# We train our model, we can make prediction

housing_prediction = tree_reg.predict(housing_prepared)

# Calculate the RMSE on the whole training set
tree_mse = mean_squared_error(housing_lables,housing_prediction)
tree_rmse = np.sqrt(tree_mse)

tree_rmse
```

No error, it's impossible ! The model must have overfit the data !

We can be sure, if we use the cross-validation (split training set into smaller pieces) : with K-fold cross-validation from the sklearn. 

We will split our training set into subsets (folds, cv=10). 
Each times we will pick a fold (1) for evaluation and training on the others (9). 

We will get an array (size = cv =10) with the evaluation scores


```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg,housing_prepared,housing_lables,
                         scoring="neg_mean_squared_error", cv=10)


tree_rmse_scores = np.sqrt(-scores)
```

Cross Validation features expect a utility function (greater is better) than a cost function (lower is better). The scoring is so the oposite of the MSE (negative value).

```python
def display(scores):
    print('Scores (RMSE) Cross Validation:\n',scores)
    print('\nMean:',scores.mean())
    print("\nStandard deviation:",scores.std())

print("The RMSE for actual prediction tree_regression:", tree_rmse,"\n")    
display(tree_rmse_scores)
```

We have a new mean for our RMSE (70666) and it's higher than  the previous one (0). So we overfit our training set. Let's compare this result with a cross validation of linear regression. 


```python
# Display the same thing for our linear regression 
scores = cross_val_score(lin_reg,housing_prepared, housing_lables,
                        scoring='neg_mean_squared_error',cv=10)

print("The RMSE for actual prediction linear regression:", lin_rmse,"\n")   
display(np.sqrt(-scores))
```

We can see that our decision tree is overfitting and it's worse than the linear regression. 
Our linear regression has no overfiting (approx. the same value for cross-validation and for individual validation) but the RMSE is too high, we still underfit.

Let's try a new complex model

### RandomForestRegressor

Training many decision trees on random subsets of the features, then average their predictions. It's an Ensemble learning.

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(housing_prepared,housing_lables)
# Our model is trained, we can now make prediction


#Compute the actual prediction's RMSE
forest_mse = mean_squared_error(housing_lables,forest_reg.predict(housing_prepared))
forest_rmse = np.sqrt(forest_mse)

print("The RMSE for actual prediction random forest:", forest_rmse,"\n")   

# We can compare it to a cross validation, to be sure we have no overfitting
forest_scores = cross_val_score(forest_reg,housing_prepared,housing_lables,
                                scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)

display(forest_rmse_scores)
```

Our model is better, we have a better mean value and std than with linear regression cross validation. 

Error on a value would be \\$50,210 +/-\\$2018

Note that we overfit our training set. We have to regularize it (add constraint with hyperparameter) or to get more data !

```python
from sklearn.base import BaseEstimator,TransformerMixin

class Only_Important_Attributes(BaseEstimator,TransformerMixin):
    
    def __init__(self,list_important_attributes=[]):
        self.list_important_attributes = list_important_attributes
    
    def fit(self,X):
        return self
    
    def transform(self,X):
        for elt in X.columns:
            if elt not in self.list_important_attributes:
                X = X.drop(elt,axis='columns')
        return X

#Crée une fonction permettant de convertir les nd_array into DF 
def to_df(nd_array,columns,ref_index):
    return pd.DataFrame(nd_array,columns=columns, index=ref_index.index)
    
print(list(housing.columns))
list_attribs = ['median_income']    
    
best_attrib = Only_Important_Attributes(list_important_attributes=list_attribs)
result = best_attrib.transform(housing)
result      
```
